### 1.2. Das erste Modell erstellen und trainieren

Es wird folgende Methoden verwendet: 
- TFIDF, 
- LogisticRegression,
- GridSearchCV 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import make_pipeline
pipe = make_pipeline(TfidfVectorizer(min_df=2, stop_words=sw, ngram_range=(1, 1), norm=None), 
                     LogisticRegression())

param_grid = {'logisticregression__C': [0.001, 0.01, 0.1, 1, 10]}                     
grid = GridSearchCV(pipe, param_grid, cv=5)
grid.fit(text_train, y_train)
print("Bester Score aus der Kreuzvalidierung: {:.2f}".format(grid.best_score_))
```

Parameter für Kosinus-Ähnlichkeit anpassen

```python
from sklearn.metrics.pairwise import linear_kernel
def find_similar(tfidf_matrix, index, top_n =999):
    cosine_similarities = linear_kernel(tfidf_matrix[index:index+1], tfidf_matrix).flatten()
    related_docs_indices = [i for i in cosine_similarities.argsort()[::-1] if i != index]
    return [(index, cosine_similarities[index]) for index in related_docs_indices][0:top_n]
(tfidf_matrix)
```

#### 4.1.2. Beim Spalten

Mit Hilfe dieses Kodes können Sie die Spaltunsposition innerhalb der Zeile finden. Die Position wird durch das Komma zwischen zwei Strings markiert. Beim Finden der Spaltungsposition ist hilfsreich darauf zu achten, dass eine linie ungefähr 25 Wörter beträgt. Der Datensatz und die Position kann man über die Variable 'anpassen' bestimmen

```python
def string_spliter(line, nte):
    first_half = ""
    last_half = ""
    for i, word in enumerate(line.split(' ')):
        if i < nte:
            first_half += word + " "
        else:
            last_half += word + " "
    return first_half.strip(" "), last_half.strip(" ")

anpassen = string_spliter(df.episode[65], 75)
print(anpassen)
```

Angepasste Angaben für den Datensatz und die Spaltungsposition eingeben und die Spaltung ausführen
