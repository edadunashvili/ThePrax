```python
import pandas as pd

def wb_updater(d, key, value):
    if key not in d:
        d[key] = [value]
    else:
        d[key] = [*d[key], value]

df = pd.read_csv('a300_string_train.csv', encoding='utf-8')

def text_rahmen(quelle, _df):
    gesep_wb={}
    for i,(einheit,gesep_full) in enumerate(zip(_df.quelle, _df.index_string)):
        gesep = (gesep_full.split('_'))[0]
        if quelle==einheit:
            wb_updater(gesep_wb,gesep,i)
    return gesep_wb

bla=text_rahmen("'300_301_302_513a_516_554_amom_22'", df)
print(bla)

def alle_text_gesep_index(_df):
    df_laenge=len(_df.index_string) 
    full_gesep_wb={}
    gesep_wb={}
    for quelle in _df.quelle:
        for i,(einheit,gesep_full) in enumerate(zip(_df.quelle, _df.index_string)):
            gesep = (gesep_full.split('_'))[0]
            if quelle==einheit:
                wb_updater(gesep_wb,gesep,i)
    return gesep_wb

print(alle_text_gesep_index(df))
```

```python
import pandas as pd
df = pd.read_csv('a300_string_train.csv', encoding='utf-8')
from collections import Counter
def ep_format(ep_full):
    return (ep_full.split('_'))[0]

def quellenvergleich(df, i1, i2):
#     print(df.quelle[i1],df.quelle[i2])
    return df.quelle[i1]==df.quelle[i2]

def unterbaum(gesep, df):
    ep_tree = {}
    a_liste = []
    z_liste = []
    
    df_len = len(df.index_string)
    for i, ep in enumerate(df.index_string):
        if gesep == ep:
#             print(i, ep_full)
            if (i > 0)&(quellenvergleich(df, i, i-1)):
                a = df.index_string[i-1]
            else:
                a = 'eAnfang_a_'
            if (i < df_len - 1):
                if not (quellenvergleich(df, i, i+1)):
                    z = 'eEnde_a_'
                else:        
                    z = df.index_string[i+1]
            else:
                z = 'eEnde_a_'
            a_liste.append(a)
            z_liste.append(z)
    return {gesep: [Counter(a_liste), Counter(z_liste)]}
```

```python
def alle_unterbaeume(gesep, df):
    episoden_baeume = {}
    ep_list = []
    for ep_full in df.index_string:
        ep = ep_format(ep_full)
        if gesep == ep:
            ep_list.append(ep_full)
    for ep in set(ep_list):
        episoden_baeume.update(unterbaum(ep, df))
            
    return episoden_baeume
```

```python
def key_val_printer(d):
    for k, v in d.items():
        print(v, k, sep=':')
    print()
        
def baum_printer(baume: dict, baum: str):
    key_val_printer(baume[baum][0])
    print(80*'-')
    print(sum(baume[baum][0].values()), baum)
    print(80*'-')
    key_val_printer(baume[baum][1])
    
def wald_printer(wald: dict):
    for baum in wald.keys():
        baum_printer(wald, baum)
        print(80*'=')
```
